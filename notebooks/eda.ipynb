# EDA — Powerpredict

Short data analysis to justify the preprocessing decisions (for the report).

```python
import os, sys
sys.path.append(os.path.abspath(os.path.join('..', 'src')))
from preprocessing import load_data, split_columns
import pandas as pd
import matplotlib.pyplot as plt

X, y = load_data()
print('X:', X.shape, '| y:', y.shape)
```

## Columns: numeric vs. text

```python
numeric, categorical = split_columns(X)
print(f'{len(numeric)} numeric columns, {len(categorical)} text columns')
print('Text columns:', categorical)
```

## Different scales -> motivates the scaling

```python
X[numeric].describe().T[['mean', 'std', 'min', 'max']]
```

## Distribution of the target (power_consumption)

```python
y.hist(bins=50)
plt.xlabel('power_consumption'); plt.ylabel('count'); plt.show()
print(y.describe())
```

## Missing values

```python
missing = X.isna().sum()
print(missing[missing > 0] if missing.any() else 'No missing values.')
```

## Strongest linear correlations with the target

```python
corr = X[numeric].join(y).corr()['power_consumption'].drop('power_consumption')
corr.abs().sort_values(ascending=False).head(10)
```
